```python
#IF STATEMENT IN PYTHON 
#Exceute Block Only IF Condition Is True 
num1=int(input("Enter Your First Number :"))
num2=int(input("Enter Your second Number :"))
if(num1>num2):
    print(f"The {num1} is greater than {num2}")
    print("This is if block.....!")
print("outside of if block ")
```

```python
#IF-ELSE STATEMENT IN PYTHON 
#USED fOR decison Making 
num1=int(input("Enter Your First Number :"))
num2=int(input("Enter Your second Number :"))
if(num1>num2):
    print(f"The {num1} is greater than {num2}")
    print("This is if block")
else:
    print(f"The {num2} is greater than {num1}")
    print("This is Else block.....!")
print("outside of if block ")
```

```python
#take 5 subjects marks from user and dispalyb total marks percentage and based on percentage display grade 
s1=int(input("Enter Your first subject marks :"))
s2=int(input("Enter Your second subject marks :"))
s3=int(input("Enter Your third subject marks :"))
s4=int(input("Enter Your fourth subject marks :"))
s5=int(input("Enter Your fifth subject marks :"))
total_marks=(s1+s2+s3+s4+s5)
max_marks=500
print(f"Total marks = {total_marks}")
per=(total_marks/max_marks)*100
print(f"Total Percentage = {per}")

if (per>90):
    print("Grade A+")
elif (per>80 and per<90):
    print("Grade A")
elif (per>70 and per<80):
    print("Grade B")
else:
    print("Grade c")
    
```

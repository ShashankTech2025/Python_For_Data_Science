METHODS OF STRING 
1. CASE CONVERSION METHODS

```python
#A=Lower
s="SHashank"
print(s.lower())
```

```python
#B=Upper
s="shashank"
print(s.upper())
```

```python
#C=Title
s="shashank bhange " # It can make only first letter capital of every word in string  
print(s.title())
```

```python
#D=Capitalize 
s="shashank bhange " # It can make only first letter capital 
print(s.capitalize())
```

```python
#E=Swapcase 
s="SHASHANK bhange" # CHange in other case like small to big and big to samll 
print(s.swapcase())
```

2. SEARCHING METHODS

```python
#A=Find
s="Python"
print(s.find("h")) #Returns index 
```

```python
s="Python"
print(s.find("z")) #Returns index or -1 if not found
```

```python
#B=Index
print(s.index("o"))
```

```python
print(s.index("z")) #Error if not found
```

```python
#C=Count 
s="Hii Hello Python "
print(s.count("l"))
```

3. STRING CHECK METHODS (BOOLEAN)

```python
s="Hello Python "
print(s.isalpha())
```

```python
#B=isdigit
s="123"
print(s.isdigit())
```

4. MODIFYING METHODS

```python
#A=replace
s="Hello"
print(s.replace("H","M"))
```

```python
#B=Strip
s=" Hii "
print(s.strip())# Removes Space 
```

```python
#C=lstrip() / rstrip()
print(s.lstrip())   # remove left space
print(s.rstrip())   # remove right space
```

5. SPLIT & JOIN (VERY IMPORTANT)

1.split() FUNCTION 


```python
#Basic Split
s = "I love Python"
print(s.split())
```

```python
#Split with custom separator
s = "a,b,c"
print(s.split(","))
```

```python
#Using maxsplit
s = "one two three four"
print(s.split(" ", 2))
```

```python
#Splitting by characters
s = "apple#banana#cherry"
print(s.split("#"))
```

2. join() FUNCTION

```python
# Basic join 
lst = ['I', 'love', 'Python']
print(" ".join(lst))
```

```python
#Custom separator
lst = ['a','b','c']
print("-".join(lst))
```

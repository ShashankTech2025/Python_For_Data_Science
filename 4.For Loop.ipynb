```python
# Break Statement 
for i in range (1,11):
    if i==4:
        break 
    print(i)
else:
    print("hello")
```

```python
# continue Statement 
for i in range (1,11):
    if i==4:
        continue 
    print(i)
else:
    print("hello")
```

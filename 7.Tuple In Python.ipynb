```python
t=("apple",20,63.25,8,6,"grapes","raj")
print(type(t))
print(t)
```

```python
# Methods Of tuple 
#1=Accessing Methods 
#A=indexing 
print(t[1])
```

```python
#B=Slicing 
print(t[0:4])
```

```python
#C= Count 
t=(1,2,2,2,2,5,3,6,55,63,5879,531)
print(t.count(2))

```

```python
#2=tuple Operations 
#A= Concatenation 
t2=(6,7,8)
print(t+t2)
```

```python
#2=Reprtition 
print(t2*2)
```

```python
#3= Checking elements 
print(6 in t)
print(656 not in t)
```

```python
#4=Built in functions 
t=(1,2,5,3,6,55,63,5879,531)
print(len(t))
print(max(t))
print(min(t))
print(sum(t))
```

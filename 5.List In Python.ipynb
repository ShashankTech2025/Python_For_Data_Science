```python
l1=["apple",20,63.25,8,6,"grapes","raj"]
print(l1)
```

#Mthods of lsit 
#1 Accesssing Elements 

```python
#Mthods of lsit 
#1 Accesssing Elements 
#a=indexing 
print(l1[0])
```

```python
#b=slicing 
#[Start:end:index]
print(l1[0:4])
```

#2 Adding Elements 

```python
#2 Adding Elements 
#A=append()--->Add element at the end of list 
l1.append(500)
print(l1)
```

```python
#B=insert()--->add element at specific position 
#l1.insert(index,value)
l1.insert(2,200)
print(l1)
```

```python
#C=extend()--->add multiple elements 
l1.extend([6,8,9])
print(l1)
```

#3 removing element 

```python
#3 removing elements 
#a= remove()--->remove value 
l1.remove(6) # when numbers are dubliacte then it can remove only first number 
print(l1)
```

```python
#B=pop()--->remove by index 
l1.pop() #remove last element 
l1.pop(1) # specific index 
print(l1)
```

```python
#C= del()---> delete element Based On index 
del l1[1]
print(l1)
```

```python
#D= clear()--> remove All in list 
l1.clear()
print(l1)
```

#4 updating elements in list 

```python
#4 updating elements in list 
l2=[10,10,10,20,30]
l2[2]=40
print(l2)
```

```python
#5  checking elements 
## in / not in
print(10 in l2)
## Count
l2.count(10)
```

```python
#6 Sorting And reversing 
l3=[23,65,89,4,7,36,5]
l3.sort()
l3
```

```python
l3=[23,65,89,4,7,36,5]
l3.reverse()
l3
```

```python
#7 Lenght and other functions 
print(len(l3))
print(max(l3))
print(min(l3))
print(sum(l3))
```

```python
#8 Looping throughout list 
for i in l3:
    print(i)
```

```python
#9 List Comprehension 
# short way to create a list 
lst=[x for x in range(1,11)]
print(lst)
```
